### Import Dependency

```python
import pandas as pd
import json
import requests
from config import client_id_yelp, api_key
```

### Perform API Call

```python
api_key = api_key
headers = {'Authorization': 'Bearer %s' % api_key}
 
url = 'https://api.yelp.com/v3/businesses/search'
params = {'term':'tacos','location':'San Diego County', 'limit':50}
 
req = requests.get(url, params=params, headers=headers)
 
parsed = json.loads(req.text)
 
businesses = parsed["businesses"]
name = []
rating = []
long = []
lat = []
review_name =[]
review_rating =[]
review_text =[]
 
for business in businesses:
    name.append(business["name"])
    rating.append(business["rating"])
    long.append(business["coordinates"]["longitude"])
    lat.append(business["coordinates"]["latitude"])
    print(business["name"]) 

    id = business["id"]
 
    url="https://api.yelp.com/v3/businesses/" + id + "/reviews"
 
    req = requests.get(url, headers=headers)
 
    parsed = json.loads(req.text)
 
    reviews = parsed["reviews"]
 
    print("--- Reviews ---")
 
    for review in reviews:
        review_name.append(review["user"]["name"])
        review_rating.append(review["rating"])
        review_text.append(review["text"])
        
        print("User:", review["user"]["name"], "Rating:", review["rating"], "Review:", review["text"], "\n")
```

### Save Reviews Tables

```python
review_df = pd.DataFrame(list(zip(review_name, review_rating, review_text)), 
               columns = ["Reviewer_name", "Reviewer_Rating", "Text"])
review_df.to_csv("reviews.csv")
```

### Save Restaurant Table

```python
df = pd.DataFrame(list(zip(name, rating, long, lat)), 
               columns = ["Restaurant", "Rating", "Longtidude", "Latitude"])
df.to_csv("locations.csv")
```
